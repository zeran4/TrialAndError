```python
import numpy as np
```

```python
data_source_row = 366
data_source_column = 6
```

```python
data_source = np.zeros((data_source_row, data_source_column), dtype=np.float32)
```

```python
print(data_source)
```

```python
for row in range(data_source_row):
    for column in range(data_source_column):
        data_source[row, column] = row + column / 10
        print("%3.1f" % data_source[row, column])
```

```python
num_of_data = 60
start_row = 0
for row in range(data_source_row - num_of_data):
    for column in range(data_source_column):
        print("start_row : ", start_row, "    sum : ", np.sum(data_source[start_row:start_row+num_of_data, column]))
        
    start_row = start_row + 1
```
